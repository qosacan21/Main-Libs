```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame ( columns=["X",'Y'])
```

```python
data
```

```python
data["X"]=np.random.normal(3,2,20)
```

```python
data["Y"]=np.random.normal(5,1.5,20)
```

```python
data
```

```python
plt.figure(figsize=(8,8))
sns.set(style="darkgrid")
sns.scatterplot(x=data["X"],y=data["Y"],color="r",markers=True)
```

```python
data.mean()
```

```python
data.hist(figsize=(10,5),)
```

```python
sns.pairplot(data)
```

```python
sns.kdeplot(data,shade=True,legend=True)
```

```python
sns.clustermap(data)
```

```python
sns.jointplot(x=data['X'],y=data['Y'],kind="kde")
```
