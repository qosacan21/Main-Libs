```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= np.linspace(0,10,100)
```

```python
x
```

```python
plt.plot(x,np.sin(x))
```

```python
plt.plot(x,np.cos(x))
plt.show()
```

```python
pd.Series(np.sin(x)).plot()
```

```python
fig = plt.figure()
plt.plot(np.sin(x),'--',color=("r"))
plt.plot(np.cos(x),'--',color=('b'))
```

```python
abc=plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.tan(x))
```
